## Clustering hiérarchique 

```python
df_clustering = df_test

#réindexer le dataframe 

#df_clustering.set_index("Product ID")

#df_clustering.drop("Product ID", axis = 1, inplace = True)
#df_clustering.drop("Last Visited", axis = 1, inplace = True)


df_clustering
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

df_cl = linkage(df_clustering,method='ward',metric='euclidean')
#df_cl = linkage(df_clustering,method='ward',metric='euclidean')


#The dendrogram illustrates how each cluster is composed 
plt.figure(figsize=(8, 5))
plt.title("AHC")
plt.xlabel('Index des échantillons')
plt.ylabel('Distance euclidienne')
dendrogram(df_cl, labels=df_clustering.index )

#dendrogram(df_cl,labels=df_clustering.index, truncate_mode="level")
plt.show()
```

```python
cluster = 3

f_clusters = fcluster(df_cl, cluster, criterion = 'maxclust')

f_clusters
```

```python
df_sample ['Cluster'] = f_clusters

cluster_1_indexes = df_sample[df_sample['Cluster'] == 1].index.tolist()
cluster_2_indexes = df_sample[df_sample['Cluster'] == 2].index.tolist()
cluster_3_indexes = df_sample[df_sample['Cluster'] == 3].index.tolist()

print("Indexes in Cluster 1:", cluster_1_indexes)
print("Indexes in Cluster 2:", cluster_2_indexes)
print("Indexes in Cluster 3:", cluster_3_indexes)

```
