##### Etape 12. Traitement des valeurs manquantes

Vous avez certainement remarqué, il y une colonne vide 'unnamed' et une ligne avec des valeurs nulles. 

Lorsque vous rencontrez des cases vides (valeurs manquantes) dans un DataFrame, vous avez plusieurs options pour les traiter :

- Suppression des lignes ou des colonnes : Vous pouvez supprimer les lignes ou les colonnes contenant des valeurs manquantes

- Remplissage des valeurs manquantes : Vous pouvez remplir les valeurs manquantes avec une valeur spécifique à l'aide de la méthode fillna() de pandas. Vous pouvez choisir de remplir avec une valeur constante, la moyenne, la médiane ou d'autres méthodes de votre choix.

- Interpolation : Vous pouvez utiliser des méthodes d'interpolation pour estimer les valeurs manquantes en fonction des valeurs voisines dans la colonne...

Le choix de la méthode dépend du contexte de vos données et de l'objectif de votre analyse. Certaines approches peuvent être plus appropriées que d'autres en fonction de la nature des données et des analyses que vous souhaitez effectuer.

###### a) Suppression des lignes ou des colonnes


```python
fr2.dropna(axis=0, inplace=True) # Voir les lignes, si on voir un null, on supprime toute la ligne
fr2.dropna(axis=1, inplace=True) # Voir les colonne, si on voir un colonne vite, supprime toute la colonne

#inplace = mettre à jour

```

###### b) Remplissage des valeurs manquantes par la moyenne

```python
moyenne_colonne = fromage_V2['calories'].mean()  
moyenne_colonne
fromage_V2['calories'].fillna(moyenne_colonne, inplace=True)
fromage_V2.tail(5)
```
